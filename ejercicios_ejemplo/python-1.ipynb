##### Ejercicio 3:
Plantear una clase Persona que contenga dos atributos: nombre y edad. Definir como responsabilidades la carga por teclado y un metodo para su impresión. Declarar una segunda clase llamada Empleado que herede de la clase Persona y agregue un atributo sueldo y muestre si debe pagar impuestos (sueldo superior a 3000)

```python
class Persona:
    def __init__(self):
        self.nombre = ""
        self.edad = 0

    def cargar_datos(self):
        self.nombre = input("Ingrese el nombre de la persona: ")
        self.edad = int(input("Ingrese la edad de la persona: "))

    def imprimir_datos(self):
      print(f"Nombre: {self.nombre} - Edad: {self.edad}")

class Empleado(Persona):
    def __init__(self):
        super().__init__()
        super().cargar_datos()
        self.sueldo = float(input("Ingrese el sueldo del empleado: "))


    def imprimir_datos(self):
        super().imprimir_datos()
        print("Sueldo:", self.sueldo)
        if self.sueldo > 3000:
            print(f"Como {self.nombre} cobra {self.sueldo} debe pagar impuestos.")
        else:
            print(f"Como {self.nombre} cobra {self.sueldo} no debe pagar impuestos.")


```

####Ejercicio 4:
Escribir una función reciba una lista de notas y devuelva la lista de calificaciones correspondientes a esas notas.

```python
def definiciones(nota):
    if nota < 5:
        calificacion = "Suspenso"
    elif nota < 6:
        calificacion = "Suficiente"
    elif nota < 7:
        calificacion = "Bien"
    elif nota < 9:
        calificacion = "Notable"
    elif nota < 10:
        calificacion = "Sobresaliente"
    else:
        calificacion = "MH"
    return calificacion

def mapeo(notas):
  return list(map(definiciones,notas))
```

```python
lista_de_notas = [4, 5.5, 7, 8.5, 10]
calificaciones_correspondientes = mapeo(lista_de_notas)
print(calificaciones_correspondientes)
```

#### Ejercicio 5:
Escribir una función que reciba una muestra de números y devuelva los valores atípicos, es decir, los valores cuya puntuación típica sea mayor que 3 o menor que -3.
Nota: La puntuación típica de un valor se obtiene restando la media y dividiendo por la desviación típica de la muestra.

```python
import numpy as np

def outliers(numeros):
  media = np.mean(numeros)
  st = np.std(numeros)
  puntuaciones = list(map(lambda numero: (numero-media)/st, numeros))
  return list(filter(lambda puntuacion: abs(puntuacion)>3, puntuaciones))
```

```python
numeros_ejemplo = [1, 2, 3, 4, 5, 100, 200, 300, 400, 500]

outliers(numeros_ejemplo)
```

```python
class Calculadora:

  def Suma(a,b):
    try:
      res = a + b
    except TypeError:
      print("Sólo puedes introducir números")
    except Exception as e:
      print("Error", type(e))
    else:
      print("Resultado: " + str(res))
    finally:
      print("Terminado")

  def Resta(a,b):
    try:
      res = a - b
    except TypeError:
      print("Sólo puedes introducir números")
    except Exception as e:
      print("Error", type(e))
    else:
      print("Resultado: " + str(res))
    finally:
      print("Terminado")

  def Multi(a,b):
    try:
      res = a * b
    except TypeError:
      print("Sólo puedes introducir números")
    except Exception as e:
      print("Error", type(e))
    else:
      print("Resultado: " + str(res))
    finally:
      print("Terminado")

  def Div(a,b):
    try:
      res = a / b
    except TypeError:
      print("Sólo puedes introducir números")
    except ZeroDivisionError:
      print("Error, no puedes dividir entre 0")
    except Exception as e:
      print("Error", type(e))
    else:
      print("Resultado: " + str(res))
    finally:
      print("Terminado")

calculadora = Calculadora
calculadora.Suma(3,4)
calculadora.Resta(3,4)
calculadora.Multi(3,4)
calculadora.Div(3,0)
```

```python
def dict_metrics(dict):
    d_serie = pd.Series(dict)
    metrics = d_serie.describe().round(2)
    return metrics[["min","max","mean","std"]]
```

##### Ejercicio 8:
Escribir un programa que genere y muestre por pantalla un DataFrame con los datos de la tabla siguiente:

Mes	Ventas	Gastos
Enero	30500	22000
Febrero	35600	23400
Marzo	28300	18100
Abril	33900	20700

Escribir una función que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas - gastos) total en los meses indicados.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
datos = {'Mes':['Enero', 'Febrero', 'Marzo', 'Abril'],
         'Ventas':[30500, 35600, 28300, 33900],
         'Gastos':[22000,23400,18100,20700]
}
df = pd.DataFrame(datos)
print("Datos: ",df)

def balance(data,lista):
  df_filt =  data[data["Mes"].isin(lista)]
  df_filt["Balance"] = df_filt["Ventas"] - df_filt["Gastos"]
  return df_filt["Balance"].sum()

print("Opción 1: ", balance(df, ["Enero","Febrero"]))

###### Opción 2:
def balance_2(datos,lista):
  datos_df = pd.DataFrame(datos)
  df_filt =  datos_df[datos_df["Mes"].isin(lista)]
  df_filt["Balance"] = df_filt["Ventas"] - df_filt["Gastos"]
  return df_filt["Balance"].sum()

print("Opción 2: ",balance_2(datos, ["Enero","Febrero"]))
```
