```python
import pandas as pd

def calculos(dicc):
  notas = pd.Series(dicc)
  estadisticas = pd.Series([notas.min(), notas.max(), notas.mean(),notas.std()], index = ["Min","Max","Media","Desviación estandar"])
  return estadisticas

diccionario = {"Paco" : 7.5,"Carlos" : 3,"Angel" : 6.8,"Clara" : 9.6,"Pedro" : 2.3,"Laura" : 5}

estadisticas = calculos(diccionario)
print(estadisticas)
```

###Apartado 1:
Carga los datos del archivo "ventas.csv" en un DataFrame de Pandas y muestra las primeras filas para revisar su estructura.

```python
import pandas as pd
ventas = pd.read_csv(r"ventas.csv")
ventas.head()
```

###Apartado 2:
Calcula el total de ingresos generados por las ventas y el número total de transacciones realizadas.

```python
ventas["ingresos"] = ventas["precio_unitario"]*ventas["cantidad"]
total_ingresos = ventas["ingresos"].sum()
print(total_ingresos)
total_transacciones = len(ventas)
print(total_transacciones)
```

```python
total_transacciones = len(ventas)
total_transacciones
```

###Apartado 3:
Identifica los productos más vendidos y muestra los cinco productos principales junto con la cantidad total vendida de cada uno.

```python
ventas.groupby("producto")["cantidad"].sum().reset_index().sort_values(by="cantidad",ascending=False)
```

###Apartado 4:
Analiza la distribución de métodos de pago utilizados.

```python
frecuencia_mpagos = ventas.groupby("metodo_pago")["producto"].count().reset_index().rename(columns={"producto":"frecuencia"}).sort_values(by="frecuencia", ascending=False)
frecuencia_mpagos.set_index("metodo_pago", inplace=True)
frecuencia_mpagos
```

```python
ventas['metodo_pago'].value_counts().plot(kind='bar',
                                            title='Distribución métodos de pago', color = "#e28743")
```

###Apartado 5:
Realiza un análisis mensual de las ventas para identificar posibles tendencias estacionales.

```python
ventas["fecha"] = pd.to_datetime(ventas["fecha"])
type(ventas.iloc[0]["fecha"])
```

```python
frecuencia_meses = ventas.groupby([ventas.fecha.dt.month, ventas.producto])["cantidad"].sum().reset_index()
frecuencia_meses.set_index("fecha",inplace=True)
frecuencia_meses = frecuencia_meses.pivot(columns ="producto")
frecuencia_meses
```

###Apartado 6:

Calcula el precio promedio y la cantidad promedio vendida por transacción. Además, identifica el producto más rentable (ingresos totales) y el menos rentable.

```python
cantidad_promedio = ventas["cantidad"].mean()
cantidad_promedio.round(2)
```

```python
precio_promedio = (ventas["precio_unitario"]*ventas["cantidad"]).mean()
precio_promedio.round(2)
```

```python
ventas_productos = ventas.groupby(["producto"]).sum()
ventas_productos["total"] = ventas_productos["cantidad"]*ventas_productos["precio_unitario"]
ventas_productos = ventas_productos.sort_values(by="total", ascending=False)
ventas_productos
```

```python
print("producto con menos ingresos: ", ventas_productos.idxmin().iloc[0],"producto con más ingresos: ", ventas_productos.idxmax().iloc[0])
```

###Apartado 7:

Analiza las ventas en función del día de la semana para identificar si existen patrones de compra específicos según el día. Calcula el total de ingresos generados y la cantidad vendida para cada día de la semana.

```python
frecuencia_diasemana = ventas.groupby(ventas.fecha.dt.weekday)[["cantidad","ingresos"]].sum().reset_index()
frecuencia_diasemana.set_index("fecha",inplace=True)
frecuencia_diasemana
```
